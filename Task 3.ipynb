## Data Visualization

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(30,30))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data)
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
data['Category'].value_counts().plot.bar(figsize=(20,10), color='blue', title='Category')
plt.subplot(222)
data['Segment'].value_counts().plot.bar(color='green', title='Segment')
plt.subplot(223)
data['Ship Mode'].value_counts().plot.bar(color='red', title='Shop Mode')
plt.subplot(224)
data['Region'].value_counts().plot.bar(color='yellow', title='Region')

plt.show()
```

```python
plt.figure(figsize=(15,10))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
sns.scatterplot("Sales",'Profit',data=data)
```

```python
pd.crosstab(data["Quantity"],data["Category"],data["Profit"],aggfunc='sum').plot(kind="bar",stacked=True)
```

```python
sns.catplot(x="Sub-Category", y="Profit", data=data, color='pink', height=4, aspect=4);
```

```python
data.plot(kind='scatter', x='Quantity', y='Profit', figsize=(12, 6), color='maroon')
plt.show()
```
